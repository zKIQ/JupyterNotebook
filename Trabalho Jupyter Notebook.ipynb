```python
#Fabíola Amaral de Azevedo Siqueira, Rafael Rezende da Silva
#Turma A, Primeiro Semestre, Análise e Desenvolvimento de Sistemas, 2018

#Cursos e notas de corte do PROUNI 2018, filtro: São José dos Campos
#https://brasil.io/dataset/cursos-prouni/cursos

#O trabalho extrai dados do link acima  
#com foco em qual o grau e turno dos cursos com corte pelo Prouni em SJC.

#com os dados se conclui que os cursos de bacharelado e a distância são os mais "ofertados"

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90

legendas = 'Bacharelado Tecnológico Licenciatura '.split()
legendas1 = 'Curso_a_Distância Noturno Matutino Integral '.split()

dados = pd.read_excel('cursos-prouni.xlsx',
                      sheet_name='cursos-prouni-75d41f9f0d454fac9')

plt.axis('equal')
plt.pie(dados['grau'].value_counts(), autopct="%.2f",
        labels=legendas, explode = [0.03,0,0])
plt.suptitle("Grau de cursos em São José dos Campos \ncom corte pelo Prouni 2018")
plt.show()
print(dados['grau'].value_counts())

print("-------------------------------------------------------------------------------")

plt.axis('equal')
plt.pie(dados['turno'].value_counts(), autopct="%.2f",
        labels=legendas1, explode = [0,0,0,0.1])
plt.suptitle("Turno de cursos em São José dos Campos \ncom corte pelo Prouni 2018")
plt.show()
print(dados['turno'].value_counts())
```
